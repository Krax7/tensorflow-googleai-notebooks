# Building the layers of the Neural Network

* En la capa de entrada (input layer), cada entrada es representada por una matriz de datos que pasarán a tavés de neuronal. Una matriz, para tensorflow es un tensor (una representación optimizada de una matriz para trabajar sobre GPU).

**Es importante entender a las entradas como cada uno de los conjuntos de características de aquello que queremos clasificar.

* Entre la capa de entrada (input layer) y la capa oculta (hidden layer), están los pesos (weights). Los pesos sob vectores que se multiplican con las matrices de entrada, de manera que, al se evaluadas por la función evaluadora, sean clasificadas de la mejor manera posible (con el error mínimo posible).

**Los pesos son "aprendidos" por al red neuronal al finalizar su entrenamiento.

* La capa oculta (hidden layer) contiene las funciones de activación que permiten evaluar la clasificación y ajustar los pesos.

* La salida (output) es la clasificación que finalmente se le dio  al conjunto de entradas que ingresaron a la red neuronal.
